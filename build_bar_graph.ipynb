```python
import matplotlib.font_manager as fm
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize=(12, 8))

# set height of bar
numerals_1 = [95.2, 89.6, 94.9, 92.2]
numerals_2 = [88.2, 81.0, 88.0, 83.4]
numerals_3 = [82.0, 71.2, 81.6, 74.9]

def neg(l):
        return [i for i in l]

number_words_1 = neg([04.8, 10.4, 05.1,  7.8])
number_words_2 = neg([11.8, 19.0, 12.0, 16.6])
number_words_3 = neg([18.0, 28.8, 18.4, 25.1])
barWidth = 0.25
br1 = np.arange(len(numerals_1))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

# Make the plot
pos_bar_1 = plt.bar(br1, numerals_1, color=mcolors.CSS4_COLORS['blue'], width=barWidth,
                    label='Words Top1')

pos_bar_2 = plt.bar(br2, numerals_2, color=mcolors.CSS4_COLORS['darkviolet'], width=barWidth, label='Words Top2')
pos_bar_3 = plt.bar(br3, numerals_3, color=mcolors.TABLEAU_COLORS['tab:red'], width=barWidth, label='Words Top3')
neg_bar_1 = plt.bar(br1, number_words_1, color=mcolors.CSS4_COLORS['cornflowerblue'], width=barWidth, bottom=numerals_1,
                    label='Numerals Top1')
neg_bar_2 = plt.bar(br2, number_words_2, color=mcolors.CSS4_COLORS['mediumorchid'], width=barWidth, bottom=numerals_2,
                    label='Numerals Top2')
neg_bar_3 = plt.bar(br3, number_words_3, color=mcolors.CSS4_COLORS['hotpink'], width=barWidth, bottom=numerals_3,
         label='Numerals Top3')
group_names = ['BERT-Base', 'RoBERTa-Base', 'BERT-Large', 'RoBERTa-Large']

ticks = [r + barWidth for r in range(len(numerals_1))]


# Create a font dictionary using the font manager
font_dict = {
    'family': 'serif',
    'size': 25,
#     'fontname': fm.findfont('Garamond')
}

# Adding Xticks
plt.xlabel('Model', **font_dict)
plt.ylabel('Percent of Examples', **font_dict)
plt.xticks(ticks, group_names, fontsize=14)
plt.yticks(fontsize=14)
font_dict = {
    'family': 'serif',
    'size': 36,
    #     'fontname': fm.findfont('Garamond')
}

# plt.title("Figure 2. Masked Tokens Numerals vs. Words ",**font_dict)
plt.legend()
plt.show()
```
